```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv('zillow.csv')
df.head()
```

```python
# Replace null values in monthly columns with their column-specific average
monthly_columns = df.columns[5:]  # Select all monthly columns (from 5th column onward)

# Replace NaN values with column-specific mean
df[monthly_columns] = df[monthly_columns].apply(lambda col: col.fillna(col.mean()), axis=0)

# Confirm that null values have been replaced
null_counts = df[monthly_columns].isnull().sum()

# Display results to verify
null_counts
```

```python
df=df.drop('StateName',axis=1)
```

```python
df=df.drop('RegionType',axis=1)
```
